```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load airline dataset 
# Drop data columns for 1985 to 1999
# Create a new dataframe called 'mod_air_data'
```

```python
air_data = pd.read_csv("C:/Lenin Data Science/DSC640/airline-safety.csv" )
mod_air_data = air_data.drop(columns = ['fatal_accidents_85_99','incidents_85_99','fatalities_85_99'])
mod_air_data
```

```python
# Create a dataframe with motor vehicle CSV file
# Extract only 3 columns Year, Fatalities and Fatality Rate
# Drop data prior to 2000 and after 2014 
# This data can be used to compare airlines data
```

```python
motor_fatal = pd.read_csv("C:/Lenin Data Science/DSC640/motorvehicle-fatalities.csv" )
mod_motor_fatal = motor_fatal[['Year','Fatalities','Fatality Rate per 100 Million VMT']]
mod_motor_fatal = mod_motor_fatal.drop(mod_motor_fatal[mod_motor_fatal.Year < 2000].index | mod_motor_fatal[mod_motor_fatal.Year > 2014].index )
mod_motor_fatal
```

```python
# Create a new dataframe 
# sum of air fatalities and 
# sum of motor fatalities
```

```python
xval = ('Air', 'Road')
yval = (new_air_data['fatalities_00_14'].sum(),mod_motor_fatal['Fatalities'].sum())
yval
```

```python
# Create a bar chart with Air Fatalities and Road(motor) Fatalities (from Year 2000-2014)
```

```python
#label locations
val = np.arange(len(xval))

# the width of the bars
bar_width = 0.20 

fig, ax = plt.subplots()
ax.bar(val + bar_width/10, yval, bar_width, color='gray')


# Create labels and Title
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Air and Road Fatalities Comparison (2000-2014)')
ax.set_xticks(val)
ax.set_xticklabels(xval, fontsize=12)

# Create labels on the bars
for i, v in enumerate(yval):
    if i == 0:
        ax.text(i-0.05,v,v, color='blue', fontsize=12)
    else:
        ax.text(i-0.05,v,v, color='#A2142F', fontsize=12, fontWeight='bold')
    
# Print a tight layout
fig.tight_layout()

#Save output
plt.savefig('C:/Lenin Data Science/DSC640/figure3.pdf', dpi=1200)
plt.show()
```

```python
# Convert kilometers into miles by Dividing with 1.6
# Multiply with 52 weeks and 15 years to calculate the number of miles travelled for 15 years
```

```python
weeks_in_year = 52
years_flown   = 15
miles_travelled = ((new_air_data['avail_seat_km_per_week']) * weeks_in_year * years_flown)/1.6
```

```python
# Calculate Fatality rate by dividing total fatalities from dataframe with total miles travelled 
# Convert rate per 100 million air miles travelled by multiplying with 100 million
# Save data to a new dataframe
```

```python
air_fatality_rate = (new_air_data['fatalities_00_14']/miles_travelled)* (10**8)
air_fatality_rate
```

```python
# Find average air fatality rate by using mean function and round to 3 decimal places
```

```python
avg_air_fatality_rate = round(air_fatality_rate.mean(),3)
avg_air_fatality_rate
```

```python
# Find average road fatality rate by using mean function and round to 3 decimal places
```

```python
avg_motor_fatality_rate = round(mod_motor_fatal['Fatality Rate per 100 Million VMT'].mean(),3)
avg_motor_fatality_rate
```

```python
# Draw bar chart between Average air fatality rate and average road fatality rate
```

```python
# label locations
yrate = [avg_air_fatality_rate, avg_motor_fatality_rate]
xrate = ['Air', 'Road']
val1 = np.arange(len(xrate))

# the width of the bars
bar_width = 0.20 

fig, ax = plt.subplots()
ax.bar(val1 + bar_width/10, yrate, bar_width, color='gray')

# Create a text box to insert on the bar chart
textstr = "The Fatality Rate ({}) for road is\n much greater than air ({})".format(avg_motor_fatality_rate, avg_air_fatality_rate)

# Create labels and Title
ax.set_ylabel('Rate(2000-2014)', fontsize=12)
ax.set_title('Fatality Rates (per 100 Million Miles Travelled)')
ax.set_xticks(val1)
ax.set_xticklabels(xrate, fontsize=12)

# Create labels on the bars
for i, v in enumerate(yrate):
    if i == 0:
        ax.text(i-0.05,v,v, color='blue', fontsize=12)
    else:
        ax.text(i-0.05,v,v, color='#A2142F', fontsize=12, fontWeight='bold')

ax.text(0.01, 0.95, textstr, fontsize=12, color='#A2142F', fontWeight= 'bold', verticalalignment='top',bbox=dict(facecolor='white', alpha=0.5))
    
# Print a tight layout
fig.tight_layout()

#Save output
plt.savefig('C:/Lenin Data Science/DSC640/figure4.pdf', dpi=1200)
plt.show()
```
